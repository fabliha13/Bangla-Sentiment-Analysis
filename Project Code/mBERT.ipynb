```python
import pandas as pd
from sklearn.model_selection import train_test_split

def numerals_normalize( df ):

  for i in range(len(df)):
    words = df['Data'][i].split()
    st = ''
    for word in words:
      if(st != ''):
        st += ' '
      if(word[0]>='0' and word[0] <='9' and word[-1]>='0' and word[-1] <='9'):
        st += "CC"
      else:
        st += word
    df['Data'][i] = st

  return df


df_train = pd.read_csv('/content/drive/MyDrive/cse424 project/Train.csv')
df_train = numerals_normalize( df_train )

df_val = pd.read_csv('/content/drive/MyDrive/cse424 project/Val.csv')
df_val = numerals_normalize( df_val )

df_test = pd.read_csv('/content/drive/MyDrive/cse424 project/Test.csv')
df_test = numerals_normalize( df_test )

df_train.to_csv(f'{directory}/Final_Train.csv', index=False)
df_val.to_csv(f'{directory}/Final_Val.csv', index=False)
df_test.to_csv(f'{directory}/Final_Test.csv', index=False)
```

```python
import os
import logging
import transformers
from transformers import BertModel, BertForSequenceClassification, BertTokenizer, AdamW, get_linear_schedule_with_warmup
import torch
import numpy as np
import pandas as pd
import seaborn as sns
from pylab import rcParams
import matplotlib.pyplot as plt
from matplotlib import rc
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from collections import defaultdict
from textwrap import wrap
from torch import nn, optim
from torch.utils.data import Dataset, DataLoader
import torch.nn.functional as F
from sklearn.utils.class_weight import compute_class_weight

# Setting up seaborn
sns.set(style='whitegrid', palette='muted', font_scale=1.2)
HAPPY_COLORS_PALETTE = ["#01BEFE", "#FFDD00", "#FF7D00", "#FF006D", "#ADFF02", "#8F00FF"]
sns.set_palette(sns.color_palette(HAPPY_COLORS_PALETTE))
rcParams['figure.figsize'] = 12, 8

# Set random seeds for reproducibility
RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
torch.manual_seed(RANDOM_SEED)
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

# Load pre-trained BERT model and tokenizer
logging.basicConfig(level=logging.ERROR)
PRE_TRAINED_MODEL_NAME = 'bert-base-multilingual-cased'
tokenizer = BertTokenizer.from_pretrained(PRE_TRAINED_MODEL_NAME)

# Dataset class
MAX_LEN = 80  # Reduced from 128
BATCH_SIZE = 64  # Same batch size

class SentnobDataset(Dataset):
    def __init__(self, reviews, targets, tokenizer, max_len):
        self.reviews = reviews
        self.targets = targets
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.reviews)

    def __getitem__(self, item):
        reviews = str(self.reviews[item])
        target = self.targets[item]

        encoding = self.tokenizer.encode_plus(
            reviews,
            add_special_tokens=True,
            max_length=self.max_len,
            return_token_type_ids=True,
            pad_to_max_length=True,
            return_attention_mask=True,
            return_tensors='pt',
        )

        return {
            'review_text': reviews,
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'targets': torch.tensor(target, dtype=torch.long)
        }

def create_data_loader(df, tokenizer, max_len, batch_size):
    ds = SentnobDataset(
        reviews=df.Data.to_numpy(),
        targets=df.Label.to_numpy(),
        tokenizer=tokenizer,
        max_len=max_len
    )
    return DataLoader(
        ds,
        batch_size=batch_size
    )

# Load data
path_parent = os.path.dirname(os.getcwd())
os.chdir(path_parent)
df_train = pd.read_csv(os.getcwd() + "/SentNoB Dataset/Final_Train.csv")
df_val = pd.read_csv(os.getcwd() + "/SentNoB Dataset/Final_Val.csv")
df_test = pd.read_csv(os.getcwd() + "/SentNoB Dataset/Final_Test.csv")

train_data_loader = create_data_loader(df_train, tokenizer, MAX_LEN, BATCH_SIZE)
val_data_loader = create_data_loader(df_val, tokenizer, MAX_LEN, BATCH_SIZE)
test_data_loader = create_data_loader(df_test, tokenizer, MAX_LEN, BATCH_SIZE)

# Model definition
class SentimentClassifier(nn.Module):
    def __init__(self, n_classes):
        super(SentimentClassifier, self).__init__()
        self.bert = BertModel.from_pretrained(PRE_TRAINED_MODEL_NAME)
        self.drop = nn.Dropout(p=0.5)
        self.out = nn.Linear(self.bert.config.hidden_size, n_classes)
        self.softmax = nn.LogSoftmax(dim=1)

    def forward(self, input_ids, attention_mask):
        last_hidden, pooled_output = self.bert(
            input_ids=input_ids,
            attention_mask=attention_mask,
            return_dict=False
        )
        output = self.drop(pooled_output)
        output = self.out(output)
        output = self.softmax(output)
        return output

model = SentimentClassifier(3)
model = model.to(device)

# Unfreeze BERT layers for fine-tuning
for name, param in model.named_parameters():
    param.requires_grad = True

# Optimizer and learning rate scheduling
optimizer = AdamW(model.parameters(), lr=2e-5, weight_decay=1e-3)
total_steps = len(train_data_loader) * 50

scheduler = get_linear_schedule_with_warmup(
    optimizer,
    num_warmup_steps=int(0.1 * total_steps),
    num_training_steps=total_steps
)

# Compute class weights
class_weights = compute_class_weight('balanced', classes=np.unique(df_train.Label), y=df_train.Label)
weights = torch.tensor(class_weights, dtype=torch.float).to(device)

loss_fn = nn.CrossEntropyLoss(weight=weights).to(device)

EPOCHS = 50

# Training and evaluation functions
def train_epoch(model, data_loader, loss_fn, optimizer, device, n_examples):
    model.train()
    losses = []
    correct_predictions = 0

    for d in data_loader:
        input_ids = d["input_ids"].to(device)
        attention_mask = d["attention_mask"].to(device)
        targets = d["targets"].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask)

        loss = loss_fn(outputs, targets)
        losses.append(loss.item())

        preds = torch.max(outputs, 1)[1]
        correct_predictions += torch.sum(preds == targets)

        optimizer.zero_grad()
        loss.backward()

        # Gradient clipping
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

        optimizer.step()
        scheduler.step()

    return correct_predictions.double() / n_examples, np.mean(losses)

def eval_model(model, data_loader, loss_fn, device, n_examples):
    model.eval()
    losses = []
    correct_predictions = 0

    with torch.no_grad():
        for d in data_loader:
            input_ids = d["input_ids"].to(device)
            attention_mask = d["attention_mask"].to(device)
            targets = d["targets"].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask)

            loss = loss_fn(outputs, targets)
            losses.append(loss.item())

            preds = torch.max(outputs, 1)[1]
            correct_predictions += torch.sum(preds == targets)

    return correct_predictions.double() / n_examples, np.mean(losses)

# Training loop with best model saving
history = defaultdict(list)
best_accuracy = 0



for epoch in range(EPOCHS):
    print(f'Epoch {epoch + 1}/{EPOCHS}')
    print('-' * 10)

    train_acc, train_loss = train_epoch(model, train_data_loader, loss_fn, optimizer, device, len(df_train))
    print(f'Train loss {train_loss} accuracy {train_acc}')

    val_acc, val_loss = eval_model(model, val_data_loader, loss_fn, device, len(df_val))
    print(f'Val   loss {val_loss} accuracy {val_acc}')

    history['train_acc'].append(train_acc)
    history['train_loss'].append(train_loss)
    history['val_acc'].append(val_acc)
    history['val_loss'].append(val_loss)

    if val_acc > best_accuracy:
        torch.save(model.state_dict(), 'best_model_state.bin')
        best_accuracy = val_acc

# Evaluate on the test set
test_acc, _ = eval_model(model, test_data_loader, loss_fn, device, len(df_test))
print(f'Test Accuracy: {test_acc.item()}')
```

```python
from sklearn.metrics import classification_report

# Step 1: Function to get predictions from the test set
def get_predictions(model, data_loader, device):
    model.eval()
    reviews = []
    predictions = []
    prediction_probs = []
    real_values = []

    with torch.no_grad():
        for d in data_loader:
            input_ids = d["input_ids"].to(device)
            attention_mask = d["attention_mask"].to(device)
            targets = d["targets"].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask)
            preds = torch.max(outputs, 1)[1]

            reviews.extend(d['review_text'])
            predictions.extend(preds)
            real_values.extend(targets)

    predictions = torch.stack(predictions).cpu()
    real_values = torch.stack(real_values).cpu()
    return predictions, real_values

# Step 2: Get predictions and true labels
y_pred, y_true = get_predictions(model, test_data_loader, device)

# Step 3: Generate the classification report
print(classification_report(y_true, y_pred, target_names=['Negative', 'Neutral', 'Positive']))
```

```python
import numpy as np
from sklearn.metrics import accuracy_score

# Function to calculate class-wise accuracy
def class_wise_accuracy(y_true, y_pred, num_classes):
    class_accuracies = {}

    for class_label in range(num_classes):
        # Get indices of the current class
        class_indices = np.where(y_true == class_label)[0]

        if len(class_indices) == 0:
            class_accuracies[class_label] = None  # No instances of this class
            continue

        # Get the true labels and predicted labels for the current class
        true_labels = y_true[class_indices]
        predicted_labels = y_pred[class_indices]

        # Calculate accuracy for this class
        accuracy = accuracy_score(true_labels, predicted_labels)
        class_accuracies[class_label] = accuracy

    return class_accuracies

# Example usage
num_classes = 3  # Change this to the number of classes in your dataset
class_accuracies = class_wise_accuracy(y_true.numpy(), y_pred.numpy(), num_classes)

# Display the class-wise accuracies
for label, accuracy in class_accuracies.items():
    print(f'Accuracy for class {label}: {accuracy:.2f}' if accuracy is not None else f'Class {label} has no instances in the dataset.')
```
